```python
# average avg_total_difftime over shares_of_buckets_with_all_sstubs_in_same_package per project

fig = plt.figure()
ax = plt.gca()

x_array = []
y_array = []

for project_info in analysis_data.values():
    x_array.append(project_info['shares_of_buckets_with_all_sstubs_in_same_package'])
    y_array.append(project_info['avg_total_difftime'])

ax.scatter(x_array, y_array, s=15)
ax.set_xlabel('shares_of_buckets_with_all_sstubs_in_same_package', fontsize=12)
ax.set_ylabel('avg_total_difftime', fontsize=12)
```

```python
# average avg_total_difftime over shares_of_buckets_with_all_sstubs_in_same_package per project

fig = plt.figure()
ax = plt.gca()

x_array = []
y_array = []
overall_diff_times = []


for project_id, project_info in analysis_data.items():
     for key, value in project_info['buckets'].items():
        overall_diff_times.append(value['thresholdDiffTime'])

for project_id, project_info in analysis_data.items():
    print('Project Name: {}'.format(project_id))
    project_x_array = []
    project_y_array = []
    all_diff_times_array = []
    interval = project_info['avg_threshold_difftime']

    for key, value in project_info['buckets'].items():
        all_diff_times_array.append(value['thresholdDiffTime'])

    project_x_array = set(all_diff_times_array)
    project_x_array = list(project_x_array)
    print('project x array: {}'.format(project_x_array))

    for unique_value in project_x_array:
        unique_value_occurs_count = all_diff_times_array.count(unique_value)
        y_percentage = unique_value_occurs_count / len(all_diff_times_array)
        project_y_array.append(y_percentage)
    print('project y array: {}'.format(project_y_array))

    print(all_diff_times_array)
    ax.plot(project_x_array, project_y_array)
```
